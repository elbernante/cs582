```python
import glob, os, pickle
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
plt.style.use('seaborn-whitegrid')
from random import shuffle
%matplotlib inline
```

```python
LOG_FILE = "output/history_log.pkl"
def get_log(filename):
    with open(filename, 'rb') as f:
        log_file = pickle.load(f)
    return log_file
```

```python
log_file = get_log(LOG_FILE)
```

```python
df = pd.DataFrame.from_dict(log_file)
df.set_index('epoch', inplace=True)
```

```python
df.tail()
```

```python
loss = df[['loss', 'val_loss']]
loss.plot(figsize=(15,5), title="Loss")
```

```python
dice = df[['dice_coef', 'val_dice_coef']]
dice.plot(figsize=(15,5), title="Dice Coefficient")
```

```python
f1 = df[['f1_score', 'val_f1_score']]
f1.plot(figsize=(15,5), title="F1 Score")
```

```python
f2 = df[['f2_score', 'val_f2_score']]
f2.plot(figsize=(15,5), title="F2 Score")
```

```python
f05 = df[['f05_score', 'val_f05_score']]
f05.plot(figsize=(15,5), title="F0.5 Score")
```
